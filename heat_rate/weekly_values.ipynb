```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("TRE_df.csv", index_col='orispl')
df
```

```python
co2_df = df.loc[:, df.columns.str.startswith('co2')]
co2_df
```

```python
# Perform calculations
co2_summary = pd.DataFrame()
co2_summary['mean'] = co2_df.mean(axis=1)
co2_summary['std'] = co2_df.std(axis=1)
co2_summary['min'] = co2_df.min(axis=1)
co2_summary['max'] = co2_df.max(axis=1)
co2_summary['range'] = co2_summary['max'] - co2_summary['min']
co2_summary
```

```python
co2_summary.describe()
```

```python
so2_df = df.loc[:, df.columns.str.startswith('so2')]
so2_summary = pd.DataFrame()
so2_summary['mean'] = so2_df.mean(axis=1)
so2_summary['std'] = so2_df.std(axis=1)
so2_summary['min'] = so2_df.min(axis=1)
so2_summary['max'] = so2_df.max(axis=1)
so2_summary['range'] = so2_summary['max'] - so2_summary['min']
so2_summary.describe()
```
